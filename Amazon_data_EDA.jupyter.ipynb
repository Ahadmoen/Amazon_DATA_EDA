# AMAZON DATASET
The dataset contains the following columns:

> **product_id:** The unique identifier for the product.

> **product_name:** The name of the product.

> **category:** The category and subcategories the product belongs to.

> **discounted_price:** The discounted price of the product.

> **actual_price:** The actual price of the product before any discounts.

> **discount_percentage:** The percentage discount applied to the product.

> **rating:** The average rating of the product.

> **rating_count:** The number of ratings the product has received.

> **about_product:** A description or details about the product.

> **user_id:** The unique identifiers of users who reviewed the product.

> **user_name:** The names of users who reviewed the product.

> **review_id:** The unique identifiers of the reviews.

> **review_title:** The titles of the reviews.

> **review_content:** The content of the reviews.

> **img_link:** The URL to the product's image.

> **product_link:** The URL to the product's page on Amazon.

### DATA CLEANING

as data types have issues and 2 rating_count values is missing so we have to clean them

```python
amazon_data['discounted_price'] = amazon_data['discounted_price'].str.replace('‚Çπ', '').str.replace(',', '').astype(float)
amazon_data['actual_price'] = amazon_data['actual_price'].str.replace('‚Çπ', '').str.replace(',', '').astype(float)
amazon_data['discount_percentage'] = amazon_data['discount_percentage'].str.replace('%', '').astype(int)
amazon_data['rating'] = pd.to_numeric(amazon_data['rating'], errors='coerce')
amazon_data['rating_count'] = amazon_data['rating_count'].str.replace(',', '').astype(float)
```

```python
amazon_data[amazon_data['rating_count'].isna()]
```

```python
cleaned_amazon_df = amazon_data.fillna(method='bfill')
```

### TREND OF DATA

**NOTE:** have to divide data into **different ranges**

##### Price Segmentation Based on Quartiles

- **Low Price Products**
  - **Criteria**: `actual_price` ‚â§ Q1 (25th percentile)
  - **Characteristics**: Represents the lower quarter of the price distribution.

- **Mid Price Products**
  - **Criteria**: Q1 < `actual_price` ‚â§ Q3 (75th percentile)
  - **Characteristics**: Represents the middle 50% of the price distribution, between the lower and upper quartiles.

- **High Price Products**
  - **Criteria**: `actual_price` > Q3 (75th percentile)
  - **Characteristics**: Represents the upper quarter of the price distribution.


```python
sns.set_style('whitegrid')
```

```python
# Calculate quartiles
Q1 = cleaned_amazon_df['actual_price'].quantile(0.25)
Q3 = cleaned_amazon_df['actual_price'].quantile(0.75)

# Segment the data
low_price_df = cleaned_amazon_df[cleaned_amazon_df['actual_price'] <= Q1]
mid_price_df = cleaned_amazon_df[(cleaned_amazon_df['actual_price'] > Q1) & (cleaned_amazon_df['actual_price'] <= Q3)]
high_price_df = cleaned_amazon_df[cleaned_amazon_df['actual_price'] > Q3]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))

# Boxplot for low-priced products
ax[0].boxplot(low_price_df['actual_price'])
ax[0].set_title('Low Price Products')
ax[0].set_xticklabels(['Low'])

ax[1].boxplot(mid_price_df['actual_price'])
ax[1].set_title('mid Price Products')
ax[1].set_xticklabels(['Mid'])

# Boxplot for high-priced products
ax[2].boxplot(high_price_df['actual_price'])
ax[2].set_title('High Price Products')
ax[2].set_xticklabels(['High'])

plt.tight_layout()
plt.show()
```
